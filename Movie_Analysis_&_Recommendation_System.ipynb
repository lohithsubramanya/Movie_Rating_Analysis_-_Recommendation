## Formulating it for other movies for a particular user

```python
import random
```

```python
movie_corr = pivot_table.corr(method='pearson', min_periods=20)
movie_corr.head()

# Assuming correlations should exist only if there are a minimum of 20 pairs of similar movies
```

```python
movie_index = random.choice(range(len(pivot_table.index)))
movie_recommendations = pd.Series()

for i in range(0, len(pivot_table.ix[movie_index].dropna().index)):
    sims = movie_corr[pivot_table.ix[movie_index].dropna().index[i]].dropna()
    
    # weighting by how much I liked the movie (inner product)
    sims = sims.map(lambda x: x*pivot_table.ix[movie_index].dropna()[i])
    movie_recommendations = movie_recommendations.append(sims)
    
total_recommendations = movie_recommendations.sort_values(ascending=False).groupby(movie_recommendations.sort_values(ascending=False).index).tail(1)

# create mask to remove duplicates
mask = total_recommendations.index.isin(pivot_table.ix[movie_index].dropna().index) 
displayed_recommendations = total_recommendations[~mask]
print (displayed_recommendations[displayed_recommendations > 0][:5])
```
